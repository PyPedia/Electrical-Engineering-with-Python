The OR gate is a fundamental logic gate used in digital electronics. It outputs True (1) if at least one of its inputs is True (1).

**1. Definition**

An OR gate performs a logical OR operation:

* The output is 1 if any input is 1.



* The output is 0 only if all inputs are 0.



**2. Truth Table**

For an OR gate with two inputs:

| Input A | Input B | Output |
| --- | --- | --- |
0ed 0an 043 | 0 | 1 | 1 |
| 1 | 0 | 1 |
| 1 | 1 | 1 |

If there are three or more inputs, the output is 1 if at least one input is 1.

**3. How It Works in Real Life**

An OR gate can be built using:

* Diodes, Transistors, or Relays.
* Integrated Circuits (ICs) for digital logic.

**4. Python Simulation**

You can simulate the OR gate in Python:|


```python
# Define an OR gate function
def or_gate(input_a, input_b):
    return input_a | input_b  # Bitwise OR operator

# Test the OR gate
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
print("Input A | Input B | Output")
for input_a, input_b in inputs:
    output = or_gate(input_a, input_b)
    print(f"   {input_a}    |    {input_b}    |    {output}")
```
