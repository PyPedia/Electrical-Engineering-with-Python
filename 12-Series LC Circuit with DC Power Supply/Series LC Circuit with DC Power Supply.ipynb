The series LC circuit with a DC power supply is an important concept in circuit theory. Here's a detailed explanation of the behavior, analysis using Kirchhoff's Voltage Law (KVL), and a numerical example:

**1. Series LC Circuit with DC Power Supply**
An LC circuit consists of an inductor (ùêø) and a capacitor (ùê∂) connected in series. When connected to a DC power supply, the behavior of the circuit depends on the transient response since the steady-state DC behavior will eventually dominate.



* Inductor (ùêø): Resists changes in current by generating a back EMF

$$v_{L}=L\frac{di}{dt}$$

* Capacitor (ùê∂): Stores energy in the electric field. The voltage across it is given by:

$$v_{C}=\frac{1}{C}\int_{}^{}idt$$

**2. Application of Kirchhoff's Voltage Law (KVL)**

For a series LC circuit with a DC source of voltage:

$$V_{dc}=v_{L}+v_{C}$$

Using the voltage expressions:

$$V_{dc}=L\frac{di}{dt}+\frac{1}{C}\int_{}^{}idt$$

This is a second-order differential equation in terms of the current ùëñ(ùë°).

**3. Circuit Analysis**

***Transient Behavior***

Initially:

* When the circuit is closed, the capacitor begins to charge, and the inductor resists the change in current.
* The current increases gradually, depending on the interplay of ùêø and ùê∂.

***Steady-State Behavior***

* In the steady state, the capacitor will be fully charged and act as an open circuit (ùëñ=0).
* The inductor behaves as a short circuit, so the voltage across it becomes zero.

**4. Solving the Differential Equation**

The differential equation for the current in the circuit is:

$$\frac{d}{dt}(V_{dc})=\frac{d}{dt}(L\frac{di}{dt}+\frac{1}{C}\int_{}^{}idt)\implies 0=L\frac{d^{2}i}{dt^{2}}+\frac{i}{C}$$

The solution to this equation is:

$$i(t)=Acos(\omega_{0}t)+Bsin(\omega_{0}t)$$

where:

$$\omega_{0}=\frac{1}{\sqrt{LC}}$$

is the *natural frequency* of oscillation of the LC circuit. The constants ùê¥ and ùêµ depend on the *initial conditions*.

**5. Numerical Example**

Given:

* Inductance ùêø=2H,
* Capacitance ùê∂=0.5 F,
* DC voltage ùëâ<sub>ùëëùëê</sub>=10 V,
* Initial current ùëñ(0)=0,
* Initial capacitor voltage ùë£<sub>ùê∂</sub>(0)=0 v.

**Step 1: Determine ùúî<sub>0</sub>**

$$\omega_{0}=\frac{1}{\sqrt{LC}}=\frac{1}{\sqrt{2\times 0.5}}=1$$

**Step 2: Solve the Differential Equation**

The general solution is:

$$i(t)=Acos(\omega_{0}t)+Bsin(\omega_{0}t)$$

Using initial conditions:

At ùë°=0, ùëñ(0)=0:

$$0=A.1+B.0\Rightarrow A=0$$

From

$$V_{dc}=v_{C}+v_{L}\Rightarrow (at \;\;t=0)\;\;\;L\frac{di}{dt}=10-0\Rightarrow \frac{di}{dt}=5$$

On the other have, we have:

$$\frac{d}{dt}(i(t))=\frac{d}{dt}(Acos(\omega_{0}t)+Bsin(\omega_{0}t))=B\omega_{0}cos(\omega_{0}t)$$

and at t=0:

$$B=5$$

**Step 3: Final Solution**

$$i(t)=5sin(t)$$

**Step 4: Voltage Across Components**

Voltage across the inductor:

$$v_{L}=L\frac{d}{dt}(5sin(t))=10cost$$

Voltage across the capacitor:

$$v_{C}=V_{dc}-v_{L}=10-10cost=10(1-cost)$$



**Python Code**

Below is a Python code to simulate and visualize the behavior of the series LC circuit described in the example:

```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
L = 2.0  # Inductance in henries
C = 0.5  # Capacitance in farads
V_dc = 10  # DC voltage in volts
omega_0 = 1 / np.sqrt(L * C)  # Natural frequency (rad/s)

# Time vector for simulation
t = np.linspace(0, 10, 1000)  # Time from 0 to 10 seconds

# Current in the circuit (solution to the differential equation)
# i(t) = 5 * sin(omega_0 * t)
i_t = 5 * np.sin(omega_0 * t)

# Voltage across the inductor: v_L = L * di/dt
v_L = L * omega_0 * 5 * np.cos(omega_0 * t)

# Voltage across the capacitor: v_C = V_dc - v_L
v_C = V_dc - v_L

# Plotting the results
plt.figure(figsize=(12, 8))

# Plot current
plt.subplot(3, 1, 1)
plt.plot(t, i_t, label="Current (i(t))", color="blue")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Current in the Series LC Circuit")
plt.grid(True)
plt.legend()

# Plot voltage across the inductor
plt.subplot(3, 1, 2)
plt.plot(t, v_L, label="Voltage across Inductor (v_L)", color="green")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Voltage Across the Inductor")
plt.grid(True)
plt.legend()

# Plot voltage across the capacitor
plt.subplot(3, 1, 3)
plt.plot(t, v_C, label="Voltage across Capacitor (v_C)", color="red")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Voltage Across the Capacitor")
plt.grid(True)
plt.legend()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
