The RC circuit in the image consists of a resistor 𝑅1=100 kΩ, a capacitor 𝐶1=1 𝜇F, and a voltage source 𝑉𝑠=10 V. I'll provide a step-by-step explanation of how to calculate the capacitor voltage 𝑉𝐶(𝑡) and the capacitor current 𝐼𝐶(𝑡) for a charging capacitor.



**Step 1: Determine the RC Time Constant (𝜏)**

The time constant for an RC circuit is given by:

$$\tau=R\times C$$

Substitute the given values:

$$\tau=R\times C=(100\times 10^{3})\times (1\times 10^{-6})=0.1s$$

**Step 2: Capacitor Voltage (𝑉<sub>𝐶</sub>(𝑡)) for Charging**

The voltage across the capacitor during charging is given by:

$$V_{c}(t)=V_{s}(1-e^{-\frac{t}{\tau}})$$

Here:

* 𝑉𝑠=10 V
* 𝜏=0.1 s

Substitute these values to get 𝑉<sub>𝐶</sub>(𝑡):

$$V_{c}(t)=10\times (1-e^{-\frac{t}{0.1}})$$

For example:

* At 𝑡=0 s, 𝑉<sub>𝐶</sub>=0
* The capacitor voltage approaches 10 V as 𝑡→∞.

**Step 4: Python Code**

Below is a Python code snippet to graph the capacitor voltage 𝑉𝐶(𝑡) and capacitor current 𝐼𝐶(𝑡) for the RC circuit:

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Vs = 10           # Voltage source (V)
R = 100e3         # Resistance (Ohms)
C = 1e-6          # Capacitance (Farads)
tau = R * C       # Time constant (seconds)

# Time range
t = np.linspace(0, 5 * tau, 500)  # 0 to 5 time constants, 500 points

# Functions
Vc = Vs * (1 - np.exp(-t / tau))  # Capacitor voltage
Ic = (Vs / R) * np.exp(-t / tau)  # Capacitor current

# Plotting
plt.figure(figsize=(12, 6))

# Plot Vc(t)
plt.subplot(2, 1, 1)
plt.plot(t, Vc, label='$V_C(t)$', color='blue')
plt.axhline(Vs, color='gray', linestyle='--', label='$V_s = 10V$')
plt.title('Capacitor Voltage and Current in RC Circuit')
plt.ylabel('Voltage (V)')
plt.grid()
plt.legend()

# Plot Ic(t)
plt.subplot(2, 1, 2)
plt.plot(t, Ic * 1e6, label='$I_C(t)$', color='red')  # Convert A to µA
plt.ylabel('Current (µA)')
plt.xlabel('Time (s)')
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
