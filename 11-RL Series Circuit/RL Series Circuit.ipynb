**Step 1: Circuit Description**

***Components:***
* Resistor ùëÖ (in ohms)
* Inductor ùêø (in henries)
* Voltage Source ùëâùë†(ùë°) (in volts)

***Series Configuration:***

* The resistor and inductor are in series, meaning the same current flows through both.



**Step 2: Kirchhoff‚Äôs Voltage Law (KVL)**

According to KVL, the sum of the voltages around a closed loop equals zero:

$$V_{s}(t)=V_{R}+V_{L}$$

Where:

$$V_{R}=i(t)\times R$$

and 

$$V_{L}=L\frac{di(t)}{d(t)}$$

and *i(t)* is the current through the circuit.

**Step 3: Differential Equation**

Substitute the expressions for ùëâ<sub>ùëÖ</sub> and ùëâ<sub>ùêø</sub> into KVL:

$$V_{s}(t)=i(t)R+L\frac{di(t)}{dt}$$

Rearranging:

$$L\frac{di(t)}{dt}+i(t)R=V_{s}(t)$$

**Step 4: Solving the Differential Equation**

If ùëâ<sub>ùë†</sub>(ùë°)=ùëâ<sub>0</sub> (a constant):

***Homogeneous Solution:***

$$L\frac{di_{h}(t)}{dt}+i_{h}(t)R=0$$

Solve for ùëñ<sub>‚Ñé</sub>(ùë°):

$$i_{h}(t)=Ae^{-\frac{R}{L}t}$$

where ùê¥ is a constant *determined by initial conditions*.

***Particular Solution:***

For a constant source, the steady-state current is:

$$i_{p}(t)=\frac{V_{0}}{R}$$

**General Solution:**

Combine the homogeneous and particular solutions:

$$i(t)=i_{p}(t)+i_{h}(t)=\frac{V_{0}}{R}+Ae^{-\frac{R}{L}t}$$

Use initial conditions (e.g., ùëñ(0)=0) to find ùê¥:

$$A=-\frac{V_{0}}{R}$$

Final solution:

$$i(t)=\frac{V_{0}}{R}(1-Ae^{-\frac{R}{L}t})$$

Hence, the current through the circuit is:



and the inductive voltage is:



**Python Code**

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
R = 4  # Ohms
L = 4e-3  # Henries (4mH)
V0 = 5  # Volts (DC source)
time = np.linspace(0, 0.01, 1000)  # Time from 0 to 10 ms

# Solution for current i(t) = V0/R * (1 - exp(-t * R / L))
i_t = (V0 / R) * (1 - np.exp(-time * R / L))

# Voltage across the inductor V_L(t) = L * di(t)/dt
d_i_t = np.gradient(i_t, time)  # Numerical derivative of i(t)
V_L_t = L * d_i_t

# Plotting
plt.figure(figsize=(10, 6))

# Plot current
plt.subplot(2, 1, 1)
plt.plot(time, i_t, label="Current (i(t))", color='b')
plt.title('Current through Circuit')
plt.xlabel('Time (s)')
plt.ylabel('Current (A)')
plt.grid(True)

# Plot voltage across the inductor
plt.subplot(2, 1, 2)
plt.plot(time, V_L_t, label="Voltage across Inductor (V_L(t))", color='r')
plt.title('Voltage across Inductor')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
