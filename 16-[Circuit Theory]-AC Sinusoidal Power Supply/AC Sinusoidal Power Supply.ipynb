Let's dive into an *AC sinusoidal power supply connected to a resistor*, step by step. We'll calculate the voltage, current, and power while understanding the mathematical principles behind them.

**1. Understanding the Circuit**

Components:

***AC sinusoidal voltage source:***
Produces a sinusoidal output, 

$$v(t)=V_{m}sin(\omega t)$$

where:

* 𝑉𝑚: Peak voltage (volts),
* 𝜔: Angular frequency, 𝜔=2𝜋𝑓,
* 𝑡: Time (seconds),
* 𝑓: Frequency (Hz).



***Resistor (𝑅):***

A purely resistive load with no phase shift.

**2. Voltage Across the Resistor**

In an AC circuit, the voltage provided by the source is:

$$v(t)=V_{m}sin(\omega t)$$

where 𝑉𝑚 is the peak voltage.

**3. Current Through the Resistor**

The current (𝑖(𝑡)) in the resistor is governed by Ohm's Law:

$$i(t)=\frac{v(t)}{R}$$

with substitution we have:

$$i(t)=\frac{V_{m}}{R}sin(\omega t)$$

define peak current as:

$$I_{m}=\frac{V_{m}}{R}$$

and so we have:

$$i(t)=I_{m}sin(\omega t)$$

**4. RMS Values**

In AC analysis, we use RMS (Root Mean Square) values:

$$V_{RMS}=\frac{V_{m}}{\sqrt{2}} \; \; , \; \; I_{RMS}=\frac{I_{m}}{\sqrt{2}}$$

**5. Power in the Resistor**

The instantaneous power (𝑝(𝑡)) is:

$$p(t)=v(t)\times i(t)$$

Substitute 𝑣(𝑡)=𝑉<sub>𝑚</sub>sin(𝜔𝑡) and 𝑖(𝑡)=𝐼<sub>𝑚</sub>sin(𝜔𝑡):

$$p(t)=V_{m}sin(\omega t)\times I_{m}sin(\omega t)=V_{m}I_{m}\frac{1-cos(2\omega t)}{2}$$

The average power is:

$$P_{avg}=\frac{V_{m}I_{m}}{2}=V_{RMS}.I_{RMS}$$

**Python Code**

The Python code calculates the voltage, current, and power for an AC circuit with a resistive load and plots the waveforms for visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
Vm = 100  # Peak voltage in volts
R = 50    # Resistance in ohms
f = 60    # Frequency in Hz
omega = 2 * np.pi * f  # Angular frequency (rad/s)

def calculate_current(Vm, R):
    Im = Vm / R  # Peak current in amps
    return Im

# Time array for one cycle
t = np.linspace(0, 1 / f, 1000)  # 1000 points within one cycle

# Voltage and current waveforms
v_t = Vm * np.sin(omega * t)  # Voltage as a function of time
Im = calculate_current(Vm, R)  # Peak current
i_t = Im * np.sin(omega * t)   # Current as a function of time

# RMS values
Vrms = Vm / np.sqrt(2)  # RMS voltage
Irms = Im / np.sqrt(2)  # RMS current

# Average power
P_avg = Vrms * Irms

# Display results
print(f"Peak Voltage (Vm): {Vm} V")
print(f"Peak Current (Im): {Im:.2f} A")
print(f"RMS Voltage (Vrms): {Vrms:.2f} V")
print(f"RMS Current (Irms): {Irms:.2f} A")
print(f"Average Power (P_avg): {P_avg:.2f} W")

# Plot voltage and current waveforms
plt.figure(figsize=(10, 6))
plt.plot(t, v_t, label="Voltage (v(t))", color="blue")
plt.plot(t, i_t, label="Current (i(t))", color="red")
plt.axhline(Vrms, color="blue", linestyle="--", label="Vrms")
plt.axhline(-Vrms, color="blue", linestyle="--")
plt.axhline(Irms, color="red", linestyle="--", label="Irms")
plt.axhline(-Irms, color="red", linestyle="--")

plt.title("Voltage and Current in an AC Circuit with Resistor")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid()
plt.show()
```
